````python
# --- Parte 4B (opcional): recomendador por género (SOLUCIÓN) ---
def top_rated_by_genre(ratings_movies, genres, top_n=10, min_ratings=50):
    """
    Filtra películas que contengan al menos uno de los géneros indicados
    y devuelve las mejor valoradas con un mínimo de valoraciones.
    """
    if isinstance(genres, str):
        genres = [genres]

    patron = "|".join(genres)

    filtrado = ratings_movies[
        ratings_movies["genres"].str.contains(patron, case=False, na=False)
    ]

    filtrado = filtrado[filtrado["num_ratings"] >= min_ratings]

    return (
        filtrado.sort_values("nota_media", ascending=False)
        .head(top_n)
        .reset_index(drop=True)
    )


# Obtener el Top 10 de películas de Acción o Drama
movies10 = top_rated_by_genre(
    ratings_movies,
    ["Action", "Drama"],
    top_n=10,
    min_ratings=50
)

# Mostrar el resultado
print(movies10)
````

````python
import pandas as pd
movies = pd.read_csv("data/movies.csv")
ratings = pd.read_csv("data/ratings.csv")
````

````python
ratings_movies = (
    ratings.groupby("movieId")
    .agg(
        nota_media=("rating", "mean"),
        num_ratings=("rating", "count")
    )
    .reset_index()
    .merge(movies, on="movieId")
)
````

````python
# Parte B.2: recommend_movies (con Gemini)
def top_rated_by_genre(df, genres, top_n=10, min_ratings=50):
    df = df[df["num_ratings"] >= min_ratings]

    mask = df["genres"].apply(
        lambda x: any(g in x for g in genres)
    )

    return df[mask].sort_values(
        by="nota_media",
        ascending=False
    ).head(top_n)
````

````python
from google import genai

client = genai.Client(api_key="GEMINI_API_KEY")  # <- pon tu clave
MODEL = "gemini-1.5-flash"
````

````python
def recommend_movies(candidates, favorite_genres, n=3):
    """
    Recomienda n películas usando Gemini SOLO del catálogo dado.
    """

    if candidates.empty:
        return []

    catalogo_texto = "\n".join(
        f"- movieId {row.movieId}: {row.title} ({row.genres}) - nota media {row.nota_media:.2f}"
        for row in candidates.itertuples(index=False)
    )

    prompt = f"""
Eres un experto recomendador de películas.

Géneros favoritos del usuario:
{", ".join(favorite_genres)}

CATÁLOGO (solo puedes elegir de aquí):
{catalogo_texto}

Selecciona exactamente {n} películas.

Devuelve SOLO un JSON válido con este formato:

[
  {{
    "movieId": 1,
    "title": "Título",
    "justificacion": "Breve explicación en español."
  }}
]

No escribas texto adicional.
"""

    try:
        response = client.models.generate_content(
            model=MODEL,
            contents=prompt
        )

        text = response.text or ""

        # limpiar markdown
        text = re.sub(r"```json|```", "", text).strip()

        # extraer JSON
        match = re.search(r"\[.*\]", text, re.DOTALL)
        if match:
            text = match.group(0)

        return json.loads(text)

    except Exception as e:
        print("Error Gemini:", e)
        return []
````
