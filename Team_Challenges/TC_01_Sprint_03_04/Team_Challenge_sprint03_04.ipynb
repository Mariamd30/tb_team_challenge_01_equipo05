````python
import pandas as pd

df_ratings = pd.read_csv('data/ratings.csv')
df_movies = pd.read_csv('data/movies.csv')

print(df_ratings.shape, df_movies.shape)
````

````python
df_pelicula_usuario_rating = pd.merge(
    df_ratings, df_movies[['movieId', 'title', 'genres']],on='movieId',how='left')
````

````python
# --- Alias de compatibilidad: la Parte 4B espera 'ratings_movies' ---

ratings_movies = df_pelicula_usuario_rating

print(ratings_movies.columns.tolist())
print(ratings_movies.head())
````

````python
# --- Parte 4B (opcional): recomendador por género (SOLUCIÓN) ---
def top_rated_by_genre(ratings_movies, genres, top_n=10, min_ratings=50):
    """
    Filtra películas que contengan al menos uno de los géneros indicados
    y devuelve las mejor valoradas con un mínimo de valoraciones.
    """
    if isinstance(genres, str):
        genres = [genres]

    patron = "|".join(genres)

    filtrado = ratings_movies[
        ratings_movies["genres"].str.contains(patron, case=False, na=False)
    ]

    # Agrupar por película: calcular nota media y número de valoraciones
    stats = (
        filtrado.groupby(["movieId", "title", "genres"])["rating"]
        .agg(nota_media="mean", num_ratings="count")
        .reset_index()
    )

    stats = stats[stats["num_ratings"] >= min_ratings]

    return (
        stats.sort_values("nota_media", ascending=False)
        .head(top_n)
        .reset_index(drop=True)
    )


# Obtener el Top 10 de películas de Acción o Drama
top_candidatos = top_rated_by_genre(
    ratings_movies,
    ["Action", "Drama"],
    top_n=10,
    min_ratings=50
)

# Mostrar el resultado
print(top_candidatos)
````

````python
# Parte B.2: recommend_movies (con Gemini)
import json

def recommend_movies(candidates, favorite_genres, n=3):
    """
    Pide a Gemini que recomiende n películas SOLO del catálogo candidato,
    con una frase de justificación en español para cada una.
    """
    catalogo = candidates[["title", "nota_media"]].to_dict(orient="records")

    prompt = f"""
Eres un recomendador de películas. Catálogo candidato:
{json.dumps(catalogo, ensure_ascii=False, indent=2)}

Géneros favoritos del usuario: {favorite_genres}

Recomienda EXACTAMENTE {n} películas SOLO de este catálogo (no inventes títulos que no estén en la lista).
Devuelve un JSON: lista de objetos con "title" y "justificacion" (frase en español).
Responde ÚNICAMENTE con el JSON, sin texto adicional ni backticks.
"""

    response = client.models.generate_content(model=MODEL, contents=prompt)
    text = response.text.strip().replace("```json", "").replace("```", "").strip()

    try:
        return json.loads(text)
    except json.JSONDecodeError:
        print("Respuesta no parseable:", text)
        return []


# Uso: recomienda a partir del Top 10 ya calculado
recomendaciones = recommend_movies(
    top_candidatos,
    favorite_genres=["Action", "Drama"],
    n=3
)
recomendaciones
````
